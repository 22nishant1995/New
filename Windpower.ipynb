```python
for i in df.columns:
  if df[i].dtype != 'object':
    plt.figure(figsize = (10,8))
    sns.boxplot(df[i])
    plt.title(f'Boxplot of {i} column')
    plt.show()
```

```python
for i in df.columns:
  if df[i].dtype != 'object':
    Q1 = df[i].quantile(0.25)
    Q3 = df[i].quantile(0.75)
    IQR = Q3-Q1
    LB = Q1-IQR*1.5
    UB = Q3+IQR*1.5
    df_new = df[(df[i]>LB) & (df[i]<UB)]

    print(f'Quantile 1{Q1} and Qantile 3 {Q3} Lower_bound is {LB} and Upper_bound is {UB}')
```
